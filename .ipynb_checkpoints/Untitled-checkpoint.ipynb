```python
def reactivity_ls_null_loc(reactivity_ls, start=5,):
    s,e=start,start
    pos_ls = []
    prev=reactivity_ls[start]
    if prev == 'NULL':
        e += 1
    for i in xrange(start+1,len(reactivity_ls)):
        if prev == 'NULL':
            if reactivity_ls[i] == 'NULL':
                e += 1
            else:
                print s,e
                pos_ls.append([s,e])
                s += 1
        else:
            if reactivity_ls[i] == 'NULL':
                s = i
                e = i+1
            else:
                s += 1
        prev = reactivity_ls[i]
    if prev == 'NULL':
#         print s,e
        pos_ls.append([s,e])
    return pos_ls
```
